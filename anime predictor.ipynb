## Stemming the Tags

```python
from nltk.stem.porter import PorterStemmer 
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return y
```

```python
mv_df['tags']=mv_df['tags'].apply(stem)
mv_df['tags']=mv_df['tags'].apply(lambda x:str(x))
```

## Converting The Tags into Vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
vec=cv.fit_transform(mv_df['tags']).toarray()
print(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

## Applying Cosine Similarity

```python
vec_a=cosine_similarity(vec)
```

```python
#for the parse function we will consider the alt title function as it has more variations of the name of the same anime. 
#This ensures the accessibility of the code even for aliases.

def recommend_2(movie):
    for z in range(len(mv_df['alt_title'])): #parse through each row by index
        for i, j in enumerate(set(mv_df['alt_title'].iloc[z])): #convert to set, give indices to match the values into i,j
            if j == movie: #compare values
                idx=z
    dist=vec_a[idx]
    anime_list=sorted(list(enumerate(dist)),reverse=True, key=lambda x:x[1])[1:6] #only top 5 values
    for i in anime_list:
        print(mv_df.iloc[i[0]].title) #to only print indexes
                
```
